```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data=pd.read_csv(r"salary_cleaned.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
colors = ['#CD6155','#5499C7','#AF7AC5','#48C9B0','#52BE80','#F4D03F']
```

```python
slices=[45, 56, 34, 20]
```

```python
plt.pie(slices);
plt.show()
```

```python
plt.pie(slices, wedgeprops = {'edgecolor':'k'})
plt.show()
```

```python
plt.pie(slices, wedgeprops = {'edgecolor':'k'},
       textprops = {'fontweight':'bold'},
       autopct = '%1.2f%%')
plt.show()
```

```python
data["workclass"].unique()
```

```python
work_count=data["workclass"].value_counts().tolist()
work_count
```

```python
work_label = data['workclass'].value_counts().index
work_label
```

```python
plt.pie(work_count, labels = work_label,
       autopct = '%1.2f%%',
       colors = colors[:3],
        wedgeprops = {'edgecolor':'k'},
       
       textprops = {'fontweight':'bold', 'size':10})
plt.show()
```

```python
plt.pie(work_count, labels = work_label,
       autopct = '%1.2f%%',
       colors = colors[:3],
       wedgeprops = {'edgecolor':'k'},
       textprops = {'fontweight':'bold', 'size':10},
       shadow = True)
plt.show()
```

```python
plt.pie(work_count, labels = work_label,
       autopct = '%1.2f%%',
       colors = colors[:3],
       wedgeprops = {'edgecolor':'k'},
       textprops = {'fontweight':'bold', 'size':10},
       shadow = True,
       explode = [0.1, 0, 0])
plt.show()
```

```python
plt.pie(work_count, labels = work_label,
       autopct = '%1.2f%%',
       colors = colors[:3],
       wedgeprops = {'edgecolor':'k'},
       textprops = {'fontweight':'bold', 'size':15},
       shadow = True,
       explode = [0.1, 0, 0],
       startangle = 90,
       pctdistance = 0.8,
        radius = 2
       )
plt.show()
```
